## 5. Preprocessament i creació del conjunt de dades per a les visualitzacions

Aquest codi defineix una funció anomenada process_chunk que pren un tros de dades com a entrada i aplica diverses operacions per preprocessar i filtrar les dades. L'objectiu final és netejar i preparar les dades per analitzar-les posteriorment.

Fem un desglossament de les principals operacions dins de la funció process_chunk:

* Excloure grups d'edat específics:
Hi ha una llista anomenada age_group_exclude que conté els grups d'edat que cal excloure de les dades. S'eliminen les files on el 'grup_edat' (grup d'edat) és a la llista d'exclusió.

* Filtrar valors específics:
Hi ha una llista anomenada values_to_filter que conté valors específics ('Impetigen', 'Escarlatina') a la columna 'diagnostic' que s'han de filtrar. S'eliminen les files on la columna 'diagnostic' coincideix amb qualsevol valor de la llista d'exclusió.

* Convertir 'data' a Datetime:
La columna 'data' es converteix en un format de data i hora.

* Filtrar interval de dates:
S'exclouen les files amb dates fora de l'interval de "2014-09-01" a "2023-08-31".

* Ordenar dades per data:
El DataFrame resultant s'ordena per la columna 'data'.

* Eliminar les columnes especificades (si s'especifica):
 Si es proporciona una llista de noms de columnes (columns_to_drop), aquestes columnes s'eliminen del DataFrame final.

La funció retorna el DataFrame preprocessat.

El codi proporciona una forma sistemàtica de netejar i preprocessar trossos de dades, especialment per a un conjunt de dades relacionades amb la vigilància de malalties infeccioses en atenció primària. Aquesta funció pot aplicar-se a trossos de dades, cosa que sol ser útil quan es tracta de grans conjunts de dades que no caben d'una vegada a la memòria.

```python
# Specify the file path
file_path = 'Vigil_ncia_sindr_mica_d_infeccions_a_Atenci__Prim_ria_20240506.csv'


# Apply operations on the chunk
age_group_exclude = ['5 a 9', '15 a 19', '70 a 74', '80 o més',
                     '10 a 14', '50 a 54', '60 a 64', '30 a 34', '45 a 49', '40 a 44',
                     '75 a 79', '20 a 24', '25 a 29', '65 a 69', '55 a 59', '35 a 39']

def process_chunk(chunk, age_group_exclude=age_group_exclude, columns_to_drop=None):


  # Remove rows based on condition
  ind = chunk[(chunk['diagnostic'] == "Bronquiolitis") & (chunk['grup_edat'].isin(age_group_exclude))].index
  chunk = chunk[~chunk.index.isin(ind)]

  # Filter out values
  values_to_filter = ['Impetigen', 'Escarlatina']
  chunk = chunk[~chunk['diagnostic'].isin(values_to_filter)]

  # Convert 'data' to datetime
  chunk['data_inici'] = pd.to_datetime(chunk['data_inici'], dayfirst=True)
  chunk['data_final'] = pd.to_datetime(chunk['data_final'], dayfirst=True)

  # Filter date range
  chunk = chunk[(chunk['data_inici'] >= "2014-09-01") & (chunk['data_inici'] <= "2023-08-31")]

  # Sort the final DataFrame by 'data'
  df = chunk.sort_values(by=['data_inici'])
  if columns_to_drop is not None:
    df = df.drop([columns_to_drop], axis=1)

  return df
```

Aquest codi té com a objectiu llegir un gran conjunt de dades a trossos, processar cada tros utilitzant la funció process_chunk (proporcionada anteriorment) i concatenar els trossos processats en un DataFrame final. L'objectiu final és obtenir un DataFrame net i preprocessat que es pugui utilitzar per a anàlisis posteriors.

Deixem una explicació pas a pas:

* Columnes a conservar:
la variable columns_to_keep especifica una llista de columnes que cal conservar del conjunt de dades original. Les guardem totes en aquest cas.

* Grandària del fragment:
la variable chunk_size especifica la mida de cada tros que es llegirà del conjunt de dades. La lectura de dades a trossos és útil quan es tracta de grans conjunts de dades que no caben a la memòria d'una sola vegada.

* Llegir i processar trossos:
el codi utilitza un bucle for per recórrer els trossos del conjunt de dades utilitzant pd.read_csv amb les columnes i la mida de tros especificats. Cada tros es processa utilitzant la funció process_chunk, que realitza operacions com filtrar, excloure valors específics i convertir tipus de dades.

* Ordenar el DataFrame final:
després de processar tots els chunks, el DataFrame final (final_df) s'ordena per la columna 'data'.

El resultat és un DataFrame (df) net i preprocessat que es pot utilitzar per a anàlisis o visualitzacions posteriors. L'ús de chunks permet al codi manejar grans conjunts de dades de forma eficient, processant-los i concatenant-los en porcions manejables.

```python
import pandas as pd

# Specify the columns you want to keep
columns_to_keep = ['setmana_epidemiologica', 'any', 'data_inici', 'data_final', 'codi_regio', 'nom_regio', 'codi_ambit', 'nom_ambit', 'codi_abs', 'nom_abs', 'diagnostic', 'sexe', 'index_socioeconomic', 'casos', 'grup_edat']

# Specify the chunk size
chunk_size = 100000

# Initialize an empty DataFrame to store the final result
final_df = pd.DataFrame()

# Read the file in chunks and process each chunk
for chunk in pd.read_csv(file_path, usecols=columns_to_keep, chunksize=chunk_size):
  # Apply the process_chunk function to each chunk
  processed_chunk = process_chunk(chunk)

  # Concatenate the processed chunk to the final DataFrame
  final_df = pd.concat([final_df, processed_chunk], ignore_index=True)

# Sort the final DataFrame by 'data_inici'
final_df = final_df.sort_values(by=['data_inici'])
df = final_df.set_index("data_inici").reset_index()
```

```python
import numpy as np
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
```

```python
# Initialize the Dash app
app = dash.Dash(__name__, )

# Define the app layout
app.layout = html.Div([
    html.H1("Disease Dashboard"),

    html.Label("Select Disease 1:"),
    dcc.Dropdown(
        id='disease-dropdown1',
        options=[{'label': disease, 'value': disease} for disease in df['diagnostic'].unique()],
        value=df['diagnostic'].unique()[0],  # Default value
        multi=False
    ),

    html.Label("Select Disease 2:"),
    dcc.Dropdown(
        id='disease-dropdown2',
        options=[{'label': 'None', 'value': 'None'}] + [{'label': disease, 'value': disease} \
                                                        for disease in df['diagnostic'].unique()],
        value='None',  # Default value
        multi=False
    ),

    dcc.Graph(id='line-plot')
])

# Define callback to update options of Disease 2 dropdown based on Disease 1 selection
@app.callback(
    Output('disease-dropdown2', 'options'),
    [Input('disease-dropdown1', 'value')]
)
def update_dropdown2_options(selected_disease1):
    if selected_disease1:
        options = [{'label': disease, 'value': disease} for disease in df['diagnostic'].unique() if \
                   disease != selected_disease1]
        return [{'label': 'None', 'value': 'None'}] + options
    else:
        return [{'label': 'None', 'value': 'None'}] + [{'label': disease, 'value': disease} for disease \
                                                       in df['diagnostic'].unique()]

# Define callback to update options of Disease 1 dropdown based on Disease 2 selection
@app.callback(
    Output('disease-dropdown1', 'options'),
    [Input('disease-dropdown2', 'value')]
)
def update_dropdown1_options(selected_disease2):
    if selected_disease2 and selected_disease2 != 'None':
        options = [{'label': disease, 'value': disease} for disease in \
                   df['diagnostic'].unique() if disease != selected_disease2]
        return options
    else:
        return [{'label': disease, 'value': disease} for disease in df['diagnostic'].unique()]

# Define callback to update the plot based on selected diseases
@app.callback(
    Output('line-plot', 'figure'),
    [Input('disease-dropdown1', 'value'),
     Input('disease-dropdown2', 'value')]
)
def update_plot(selected_disease1, selected_disease2):
    if not selected_disease1 and selected_disease2 == 'None':
        return {}

    selected_data1 = df[df['diagnostic'] == selected_disease1] if selected_disease1 else pd.DataFrame()
    selected_data2 = df[df['diagnostic'] == selected_disease2] if \
    selected_disease2 and selected_disease2 != 'None' else pd.DataFrame()

    combined_data = pd.concat([selected_data1, selected_data2])

    daily_cases = combined_data.groupby(['data_inici', 'diagnostic']).sum('casos').reset_index()

    traces = []

    for i, selected_disease in enumerate([selected_disease1, selected_disease2]):
        if selected_disease and selected_disease != 'None':
            trace = {
                'x': daily_cases[daily_cases['diagnostic'] == selected_disease]['data_inici'],
                'y': daily_cases[daily_cases['diagnostic'] == selected_disease]['casos'],
                'type': 'line',
                'name': selected_disease,
                'yaxis': f'y{i+1}' if i == 1 else 'y'
            }
            traces.append(trace)

    layout = {
        'title': f'Daily Cases of {selected_disease1} and {selected_disease2}' \
        if selected_disease2 and selected_disease2 != 'None' else f'Daily Cases of {selected_disease1}',
        'xaxis': {
            'title': 'Date',
            'tickvals': daily_cases['data_inici'].dt.strftime('%Y-%m').unique()[::6],
            'ticktext': daily_cases['data_inici'].dt.strftime('%Y-%m').unique()[::6],
            'tickfont': {'size': 9},
            'tickangle': 90  # Adjust the tick angle as needed
        },
        'yaxis': {'title': f'Cases ({selected_disease1})'},
        'yaxis2': {'title': f'Cases ({selected_disease2})', 'overlaying': 'y', 'side': 'right'} if \
        selected_disease2 and selected_disease2 != 'None' else {}
    }

    return {'data': traces, 'layout': layout}

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True, port=8050)
```

La gràfica mostra que hi ha una clara estacionalitat a gairebé totes les malalties.
Casos alts: el pic de casos es registra al voltant de febrer de cada any.
Casos baixos: al començament de cada temporada (setembre) es registren els casos més baixos.

Efecte COVID: al pic de COVID, veiem que la tendència s'interromp, però es restableix a principis de $2023$.
